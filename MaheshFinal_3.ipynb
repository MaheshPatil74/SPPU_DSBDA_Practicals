```python
def display_stats(name,data):
    print(f'\n {name} Statistics : ')
    print('\nMean : ',data.mean( numeric_only=True) )
    print('\nMedian : ',data.median(numeric_only=True) )
    print('\nStandard Deviation : ',data.std(numeric_only=True) )   
    print('\nPercentiles :\n', data.quantile([0.25,0.50,0.75],numeric_only=True) )
```

```python
display_stats('Setosa', setosa)
display_stats('Versicolor', versicolor)
display_stats('Virginica', virginica)
```
