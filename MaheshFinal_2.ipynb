```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Name' : [ 'A' , 'B' , 'C' , 'D' , 'E' , 'F' , 'G' , 'H' ],
    'English': [ 85, 90, np.nan, 40, 100, 95, 200, 60 ],
    'Science': [80, np.nan, 70, 45, 90, 85, 88, 300 ],
    'Math' : [75, 85, 65, np.nan, 95, 90, 92, 55],
    'Attendance' : [90, 95, 85, 60, np.nan, 98, 40, 88],
    'Gender' : [ 'Male' , 'm' , 'male' , 'Male' , 'Male' , 'm' , 'male' , 'Male' ]    
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['English'] = df['English'].fillna( df['English'].mean() )
df['Science'] = df['Science'].fillna( df['Science'].mean() )
df['Math'] = df['Math'].fillna( df['Math'].mean() )
df['Attendance'] = df['Attendance'].fillna( df['Attendance'].mean() )
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['English'] = df['English'].clip(0,100)
```

```python
df['Gender'] = df['Gender'].str.lower()
df['Gender'] = df['Gender'].replace({
    'male' : 'Male',
    'm' : 'Male'
})
```

```python
df
```

```python
Q1 = df[['English','Science','Math','Attendance']].quantile(0.25)
Q3 = df[['English','Science','Math','Attendance']].quantile(0.75)
print('Q1 : \n',Q1)
print('Q3 : \n',Q3)
```

```python
IQR = Q3-Q1
outlier = df[( (df[['English','Science','Math','Attendance']] < (Q1 - 1.5*IQR)) | (df[['English','Science','Math','Attendance']] > (Q3 + 1.5*IQR)) )]
outlier
```

```python
df[['English','Science','Attendance']] = df[['English','Science','Attendance']].clip(70,100)
```

```python
df
```

```python
df['Attendance_log'] = np.log(df['Attendance']+1)
df
```

```python
df['Attendance_sq'] = np.sqrt(df['Attendance']+1)
df
```

```python
df['Attendance_log'] = 1/(df['Attendance']+1)
df
```
