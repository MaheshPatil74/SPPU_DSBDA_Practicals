```python
import nltk
from nltk.tokenize import sent_tokenize , word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer , WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = """
Data Science is an interdisciplinary field.
It uses scientific methods, processes,
algorithms and systems to extract knowledge
from data.
"""
print("Original Text : ",text)
```

```python
# Tokenization
```

```python
sentence_tokens = sent_tokenize(text)
print("\nSentence Tokens : " , sentence_tokens)
```

```python
tokens = word_tokenize(text)
print("\nWord Tokenize : " , tokens)
```

```python
# pos tagging
```

```python
pos_tags = pos_tag(tokens)
print("\POS Tags : " , pos_tags)
```

```python
# Stop Words Removal
```

```python
stop_words = set( stopwords.words('english'))
filtered_words = []
for word in tokens :
    if word.isalpha() and word.lower() not in stop_words :
        filtered_words.append(word)
print('\nFiltered Words : ', filtered_words)
```

```python
# Stemming
```

```python
stemmer = PorterStemmer()
stemmed_words = []
for word in filtered_words :
    stemmed_words.append( stemmer.stem(word) )
print('\nStemmed_Words : ', stemmed_words)
```

```python
# Lemmatization
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = []
for word in filtered_words :
    lemmatized_words.append( lemmatizer.lemmatize(word) )
print('\nLemmatized Words : ', lemmatized_words)
```

```python
# TF-IDF representation
```

```python
documents = [

    "Data science uses machine learning",

    "Machine learning extracts patterns from data",

    "Data science and analytics are related"
]
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
print("\nTF-IDF Feature Name : ", vectorizer.get_feature_names_out() )
print("\nTF-IDF Matrix : ", tfidf_matrix.toarray() )
```
