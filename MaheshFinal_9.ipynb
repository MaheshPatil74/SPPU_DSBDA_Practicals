```python
# Main Boxplot
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=df,x='sex',y='age',hue='survived')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Age Distribution by gender and Survival')
plt.show()
```

```python
sns.countplot( data=df , x='survived' , hue='sex' , palette='seismic' )
plt.title('Survival Count by Gender')
plt.show()
```

```python
sns.countplot( data=df , x='sex' , hue='survived' , palette='seismic' )
plt.title('Survival Count by Gender')
plt.show()
```

```python
sns.histplot(data=df , x='age' , hue='sex', kde=True , palette='seismic' )
plt.title('Age Distribution by Gender')
plt.show()
```
